```python
import pandas as pd
from googletrans import Translator

def translate_to_tamil(text):
    translator = Translator()
    translated_text = translator.translate(text, src='en', dest='ta').text
    return translated_text

def translate_csv_to_tamil(input_file, output_file):
    df = pd.read_csv(input_file)
    df['Tamil_Name'] = df['names'].apply(translate_to_tamil)
    df.to_csv(output_file, index=False, encoding='utf-8-sig')

if __name__ == "__main__":
    input_file = 'input.csv'
    output_file = 'output.csv'
    translate_csv_to_tamil(input_file, output_file)
```
